```python
import numpy as np   #To do statistical analysis of data
```

```python
""""To check the presence of outliers. 
    Outliers are data points that are very different from the rest of the data.
    They stand out because they're much higher or lower than most other points."""
outliers=[]
age=x1['age']

def detect_outliers_iqr(df):
    df = sorted(df)
    q1 = np.percentile(df,25)
    q3 = np.percentile(df,75)
    
    iqr = q3-q1
    print(q1,q3)
    lwr_bound = q1-(1.5*iqr)
    upr_bound = q3+(1.5*iqr)
    print(lwr_bound,upr_bound)
    
    for i in df:
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers

marks_outliers = detect_outliers_iqr(age)
print("outliers",marks_outliers)
print(outliers)
```
